# 9.6 Consultas com INNER JOIN e ORDER BY

```
pd.read_sql_query('''
SELECT ALUNO.nome, CHAMADO.descricao, CHAMADO.dt_abertura, CHAMADO.dt_fechamento
FROM ALUNO
INNER JOIN CHAMADO
ON ALUNO.id = CHAMADO.FK_ALUNO_ID
ORDER BY CHAMADO.dt_abertura ASC
''', conn)
```

```
pd.read_sql_query('''
SELECT ALUNO.matricula as "matricula aluno", ALUNO.nome as aluno, ORIENTADOR.matricula as "matricula orientador", ORIENTADOR.nome as orientador
FROM ALUNO
INNER JOIN ORIENTADOR
ON ALUNO.FK_ORIENTADOR_id = ORIENTADOR.id
WHERE ALUNO.status = 'Ativo'
ORDER BY ORIENTADOR.nome ASC
''', conn)
```

```
pd.read_sql_query('''
SELECT AGENTE.nome as agente, EMPRESA.nome as empresa
FROM AGENTE
INNER JOIN EMPRESA
ON AGENTE.FK_EMPRESA_id = EMPRESA.id
ORDER BY AGENTE.nome ASC
''', conn)
```

```
pd.read_sql_query('''
SELECT ALUNO.nome as aluno, AGENTE.nome as agente
FROM ALUNO
INNER JOIN AGENTE
ON ALUNO.FK_AGENTE_ID = AGENTE.id
ORDER BY ALUNO.NOME ASC
''', conn)
```

```
pd.read_sql_query('''
SELECT ALUNO.matricula as matricula, ALUNO.nome as aluno, CHAMADO.descricao as chamado, CHAMADO.dt_abertura as data_abertura
FROM ALUNO
INNER JOIN CHAMADO
ON ALUNO.id = CHAMADO.FK_ALUNO_id
WHERE ALUNO.status = 'Inativo' AND CHAMADO.status = 'Aberto'
ORDER BY CHAMADO.dt_abertura ASC
''', conn)
```

```
pd.read_sql_query('''
SELECT SERVIDOR.nome as Servidor, CHAMADO.descricao , ALUNO.nome as Aluno, ORIENTADOR.nome as Orientador, AGENTE.nome as Agente, EMPRESA.nome as Empresa
FROM SERVIDOR
INNER JOIN CHAMADO
ON SERVIDOR.id = CHAMADO.FK_SERVIDOR_ID
INNER JOIN ALUNO
ON CHAMADO.FK_ALUNO_ID = ALUNO.id
INNER JOIN EMPRESA
ON ALUNO.FK_EMPRESA_ID = EMPRESA.ID
INNER JOIN AGENTE
ON ALUNO.FK_AGENTE_ID = AGENTE.ID
INNER JOIN ORIENTADOR
ON ALUNO.FK_ORIENTADOR_ID = ORIENTADOR.ID
ORDER BY SERVIDOR.nome DESC
''', conn)
```

# 9.9 CONSULTAS COM SELF JOIN E VIEW

```
pd.read_sql_query('''
SELECT ALUNO1.matricula as "matricula 1", ALUNO1.nome as "nome 1", ALUNO2.matricula as "matricula 2", ALUNO2.nome as "nome 2", empresa.nome as "empresa"
FROM ALUNO ALUNO1
INNER JOIN ALUNO ALUNO2 ON
ALUNO1.FK_EMPRESA_id = ALUNO2.FK_EMPRESA_id AND ALUNO1.id < ALUNO2.id
INNER JOIN EMPRESA ON
ALUNO1.FK_EMPRESA_id = EMPRESA.id
ORDER BY EMPRESA.nome ASC, ALUNO1.nome ASC, ALUNO2.nome ASC
''', conn)
```

```
pd.read_sql_query('''
CREATE VIEW joninho AS
SELECT nome, matricula, status
FROM ALUNO
WHERE FK_AGENTE_ID = 1
''', conn)
```

```
pd.read_sql_query('''
CREATE VIEW orientador_aluno AS
SELECT ORIENTADOR.matricula as "matricula orientador", ORIENTADOR.nome as orientador, ALUNO.matricula as "matricula aluno", ALUNO.nome as aluno
FROM ORIENTADOR
INNER JOIN ALUNO
ON ORIENTADOR.id = ALUNO.FK_ORIENTADOR_id
''', conn)
```

```
pd.read_sql_query('''
CREATE VIEW fechados AS
SELECT descricao, dt_abertura, dt_fechamento
FROM CHAMADO
WHERE status = 'Fechado'
''', conn)
```

```
pd.read_sql_query('''
CREATE VIEW chamados_abertos AS
SELECT ALUNO.matricula as matricula, ALUNO.nome as aluno, CHAMADO.descricao as chamado, CHAMADO.dt_abertura as data_abertura, SERVIDOR.nome as servidor
FROM ALUNO
INNER JOIN CHAMADO
ON ALUNO.id = CHAMADO.FK_ALUNO_id
INNER JOIN SERVIDOR
ON CHAMADO.FK_SERVIDOR_id = SERVIDOR.id
WHERE CHAMADO.status = 'Aberto'
''', conn)
```

```
pd.read_sql_query('''
CREATE VIEW miranda AS
SELECT descricao, status, dt_abertura, dt_fechamento
FROM CHAMADO
WHERE FK_SERVIDOR_ID = 3
''', conn)
```

# **9.10 Subconsultas**

```
pd.read_sql_query('''
SELECT aluno.nome AS nome,
(SELECT nome FROM ORIENTADOR WHERE id = aluno.fk_orientador_id) AS orientador_nome,
(SELECT nome FROM AGENTE WHERE id = aluno.fk_agente_id) AS agente_nome,
(SELECT nome FROM EMPRESA WHERE id = aluno.fk_empresa_id) AS empresa_nome
FROM ALUNO
WHERE status = 'Ativo'
''', conn)
```

```
pd.read_sql_query('''
SELECT
(SELECT nome FROM ALUNO WHERE id = chamado.fk_aluno_id) AS aluno_nome,
(SELECT nome FROM SERVIDOR WHERE id = chamado.fk_servidor_id) AS agente_nome,
descricao
FROM CHAMADO
''', conn)
```

```
pd.read_sql_query('''
SELECT nome AS aluno_nome, total_chamados
FROM (SELECT aluno.nome AS nome, COUNT(chamado.id) AS total_chamados FROM ALUNO
     JOIN CHAMADO ON aluno.id = chamado.fk_aluno_id
     WHERE chamado.status = 'Aberto'
     GROUP BY aluno.nome
     HAVING COUNT(chamado.id) > 1) AS alunos_com_chamados_abertos
''', conn)
```

```
pd.read_sql_query('''
SELECT SERVIDOR.nome, COUNT(CHAMADO.id) as "qtd chamados"
FROM SERVIDOR
INNER JOIN CHAMADO
ON SERVIDOR.id = CHAMADO.FK_SERVIDOR_id
GROUP BY SERVIDOR.id
HAVING COUNT(CHAMADO.id) > (
    SELECT AVG(total) as media
    FROM (
        SELECT COUNT(*) as total
        FROM CHAMADO
        GROUP BY FK_SERVIDOR_id
    ) as media_chamados
)
''', conn)
```

```
pd.read_sql_query('''
SELECT AVG(total) as media
FROM (
    SELECT COUNT(*) as total
    FROM CHAMADO
    GROUP BY FK_SERVIDOR_id
) as media_chamados
''', conn)
```
